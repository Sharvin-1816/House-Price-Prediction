```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv(r"D:\DOWNLOADS\house-prices-advanced-regression-techniques\train.csv")
```

```python
df
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)
```

```python
df[categorical_columns] = df[categorical_columns].fillna('missing')
```

```python
encoder = OrdinalEncoder()
df[categorical_columns] = encoder.fit_transform(df[categorical_columns])
```

```python
df
```

```python
categorical_columns
```

```python
l=[]
for x in df.columns:
    if x not in categorical_columns:
        l.append(x)
```

```python
l
```

```python
scaler = StandardScaler()
df[l] = scaler.fit_transform(df[l])
```

```python
df
```

```python
df[categorical_columns] = scaler.fit_transform(df[categorical_columns])
```

```python
df.fillna(0, inplace=True)
```

```python
last_column_name = df.columns[-1]
df[last_column_name] = scaler.fit_transform(df[[last_column_name]])
```

```python
df
```

```python
X = df.iloc[:,1:len(df.columns)-1]
Y = df.iloc[:,-1]
```

```python
# X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```
